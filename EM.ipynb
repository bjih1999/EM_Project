```python
import csv
import numpy as np
import random
from matplotlib import pyplot as plt
from matplotlib import mlab
from scipy.stats import multivariate_normal

random.seed(1)

def getGraph(mean, cov):

    # plt.figure(figsize=(10, 4))
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    for index in range(K):
        x, y = np.mgrid[-20:140, -20:140]
        pos = np.dstack((x, y))
        rv = multivariate_normal(mean[index], cov[index])
        ax.plot_surface(x, y, rv.pdf(pos), alpha=0.8, cmap='viridis')

    plt.figure()
    ax.view_init(elev=90)
    plt.figure()
    # plt.subplot(2, 1, 2) 

def getPlot(gaussian_set):

    for gaussian in gaussian_set:
        gaussian = np.array(gaussian)
        plt.scatter(gaussian[:, 0], gaussian[:, 1])
    
    plt.show()

if __name__ == '__main__':
    
    K = 3
    N = 1800
    count = 0

    # read file
    points = []
    with open('./points.csv') as file:
        point_info = csv.reader(file)

        for point in point_info:
            points.append(list(map(float, point)))

    # initialize
    gaussian_set = [[], [], []]
    expected_in = [[0 for _ in range(K)] for _ in range(N)]
    mean = []
    cov = []
    weight = []
    for index, point in enumerate(points):
        j = random.randint(0, 2)
        gaussian_set[j].append(point)
        expected_in[index][j] = 1

    for index, gaussian in enumerate(gaussian_set):
        mean.append(np.mean(gaussian, axis=0))
        cov.append(np.cov(np.array(gaussian).T))
        weight.append(len(gaussian) / N)
    
    # getGraph(mean, cov)
    # getPlot(gaussian_set)

    # for index, point in enumerate(points):
    #     j = random.randint(0, 2)
    #     gaussian_set[j].append(point)
    #     expected_in[index][j] = 1

    # for index, gaussian in enumerate(gaussian_set):
    #     mean.append(np.mean(gaussian, axis=0))
    #     # print('transposed : ', np.array(gaussian).T)
    #     cov.append(np.cov(np.array(gaussian).T))
    #     weight.append(len(gaussian) / N)
    
    print(mean)
    print(cov)
    print(weight)
    getGraph(mean, cov)
    getPlot(gaussian_set)


    new_expect = [[0 for _ in range(K)] for _ in range(N)]

    while(expected_in != new_expect):
        expected_in = new_expect
        count += 1

        # Expectation
        rvs = []
        new_expect = [[0 for _ in range(K)] for _ in range(N)]
        for index in range(K):
            # print(mean[index])
            # print(cov[index])
            rvs.append(multivariate_normal(mean[index], cov[index]))

        gaussian_set = [[], [], []]
        for i in range(N):
            whole_probability = 0
            probabilities = []
            for k in range(K):
                probabilities.append(weight[k] * rvs[k].pdf(points[i]))

            whole_probability = sum(probabilities)

            highest = -1
            gaussian_index = 0
            for j in range(K):
                probability = probabilities[j]/whole_probability
                if highest < probability:
                    highest = probability
                    gaussian_index = j

            gaussian_set[gaussian_index].append(points[i])
            new_expect[i][gaussian_index] = 1

        # Maximization
        mean = []
        cov = []
        weight = []
        for index, gaussian in enumerate(gaussian_set):
            mean.append(np.mean(gaussian, axis=0))
            cov.append(np.cov(np.array(gaussian).T, bias=True))
            weight.append(len(gaussian) / N)
        
        if count % 5 == 0:
            print(mean)
            print(cov)
            print(weight)
            
            getGraph(mean, cov)
            getPlot(gaussian_set)

    print(count)
    plt.show()



```
